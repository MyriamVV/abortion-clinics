```python
import pandas as pd
```

```python
# The abortion-distances database contains information about the 
# distances women from all of the counties of the U.S. have to travel
# to get to an abortion facility. It's a monthly analysis 
# from 2009 to 2021

abortion = pd.read_csv("abortion-distances.csv")
abortion
```

```python
abor_new = abortion[["origin_county_name", "monthlydate", "year", 
                     "origin_county_pop", "gd", "gd_city"]]
abor_new.sample(10)
```

```python
# I decided to sort the information to end up with only the 
# first month of 2021
abor_new_sort = abor_new.sort_values(by=["monthlydate"], ascending = True)
abor_new_sort
```

```python
# I used a mask to extract the information  
abor_new_sort['monthlydate'].eq('2021m1')
```

```python
# This applies the mask to the database 
abor_new_sort[abor_new_sort['monthlydate'].eq('2021m1')]
```

```python
# Here I made the sorting permanent

abor_new_sort = abor_new_sort[abor_new_sort['monthlydate'].eq
                              ('2021m1')]
abor_new_sort
```

```python
abor_new_sort.to_csv("abortion-final.csv", encoding = "UTF-8", index = False)
```

```python
abortion = pd.read_csv("abortion-final.csv")
abortion
```

```python
# The next step is figuring out how many counties are in the csv
# This is the first step to figure out the minimum and maximum
# distances women have to travel, what are the destination cities
# and from which counties do they come from
abortion["gd_city"].unique()
```

```python
# First turn the year into an object 

abortion['year'] = abortion['year'].apply(pd.to_datetime)
```

```python
# To figure out how many people rely in a city, I need a group-by
# command 

abortion.groupby('gd_city').sum()
```

```python
# Lets filter it to eliminate gd

abort_population=abortion.groupby('gd_city').sum()[['origin_county_pop']].reset_index()
abort_population
```

```python
# Now transform the soted data into a csv

abort_population.to_csv("abort_population_total.csv", index = False)
```

```python
# To figure out the distances, min and max. This can be eliminated

abortion.groupby('gd_city').max()
```

```python
abortion.groupby('gd_city').min()
```

```python
# Another alternative to sorting the distances is paring the city 
# with max distance. 
# Maddy explained to me that if the names are bolded, it means 
# I can't use the information. 
abortion.groupby('gd_city').max()[['gd']]
```

```python
# But we want to use it, so we use "reset index"
abortion.groupby('gd_city').max()[['gd']].reset_index()
```

```python
# We asigned the changes. Elminating unnecesary columns 

cities = abortion[['origin_county_name', 'origin_county_pop', 'gd', 'gd_city']]
```

```python
# I group by it again. To find the distance for each destination city.
# This is giving me biggests distance to get to the city. 
max_distance = abortion.groupby('gd_city').max()[['gd']].reset_index()
max_distance
```

```python
# Now we merge the two lists. To find the origin county 
# name and population

max_dis_merged = max_distance.merge(cities, on = ["gd", "gd_city"])
```

```python
# For the minimum distance

min_distance = abortion.groupby('gd_city').min()[['gd']].reset_index()
min_distance
```

```python
# Merge again

min_dis_merged = min_distance.merge(cities, on = ["gd", "gd_city"])
```

```python
# index = False. This eliminates the weird left numbers in the
# colum
min_dis_merged.to_csv("min-distances.csv", index = False)
```

```python
max_dis_merged.to_csv("max-distances.csv", index = False)
```

```python
# Let's try comparing them in pandas. Sort them

abortion_min_merged = pd.read_csv("min-distances.csv")
abortion_min_merged

```
